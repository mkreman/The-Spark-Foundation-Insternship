# State Wise Analysis of Sales, Discount and Profit

```python
data['State'].value_counts()
```

```python
data_state = data.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
data_state
```

```python
plt.figure(figsize = (20,20))
plt.pie(data_state['Sales'], labels = data_state.index, autopct = '%1.1f%%')
plt.title('State Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

- **The highest amount(11.8%) of sales is in Wyoming State**
- **The lowest amount(0.8%) of sales is in South Dakota state**

```python
data_state.sort_values('Profit')[[ 'Profit']].plot(kind = 'bar', figsize = (15,10))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

 - Vermont at the Top
 - Ohio at the Bottom

```python
data_state.sort_values('Discount')[[ 'Discount']].plot(kind = 'bar', figsize = (15,10))
plt.title('State wise Discount Analysis', fontsize = 20)
plt.ylabel('Discount per Sate')
plt.xlabel('States')
plt.show()
```

Illinois at the top
